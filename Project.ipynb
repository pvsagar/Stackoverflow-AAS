### 6. Tags Scraping Script and Plot

* Scraping first 50 pages
* Creating tags_dataset.csv
* Plotting Count by tag

```python
# Function to get the Tag Data
def getTagData():               
    link=None
    tags=[]                                               
    counts=[]
    
    page_url="https://stackoverflow.com/tags?page=1&tab=popular"  
    
    for i in range(50):
        #Sends a get request to the web page
        page=requests.get(page_url)
        #Status_code 200 indicates success
        if page.status_code==200:
            page.content
            
            #Initiates a beautifulsoup object using the html source and Python’s html.parser
            soup=BeautifulSoup(page.content, 'html.parser')
            
            #Structures soup object properly
            soup.prettify()                                   

            divs=soup.select("div.grid-layout--cell")
                                
            for idx, div in enumerate(divs):
                tag=div.select("a.post-tag")[0].get_text()
                count=div.select("span.item-multiplier-count")[0].get_text()
                count=int(count)
                if(tag):
                    tags.append(tag)
                else:
                    tags.append(np.nan)
                    
                if(count):
                    counts.append(count)
                else:
                    counts.append(np.nan)

                next_page=soup.select("div.pager a")


            for idx, item in enumerate(next_page): 
                if item.select("span.page-numbers")[0].get_text()==" next":
                    link=item.get('href')
                    break
                else:
                    continue

            if not link:       
                break         
            page_url = urljoin("https://stackoverflow.com",link)
    
    tagsdf = pd.DataFrame(tags)
    countsdf = pd.DataFrame(counts)
    
    dataframe = pd.concat([tagsdf,countsdf], axis=1, sort=False)
    
    dataframe.columns = ['Tag', 'Tag_Count']
    dataframe.to_csv("tags_dataset.csv", sep=',', index=False, encoding='utf-8') 
    print("Tags data saved in tags_dataset.csv in the current working directory\n")
    return dataframe

if __name__=="__main__":
    tags_data=getTagData()
```

### 7. Users Profile Scraping Script and Plot

* Scraping first 50 pages
* Creating users_dataset.csv
* Plotting Reputation Score by user

```python
# Function to get the User Data
def getUserData():               
    link=None
    usernames=[]
    locations=[]                                               
    reputation_score=[]
    
    page_url="https://stackoverflow.com/users?tab=Reputation&filter=all"  
    
    for i in range(50):
        #Sends send a get request to the web page 
        page=requests.get(page_url)
        #Status_code 200 indicates success 
        if page.status_code==200:
            page.content                      
            
            #Initiates a beautifulsoup object using the html source and Python’s html.parser
            soup=BeautifulSoup(page.content, 'html.parser')
            
            #Structures soup object properly
            soup.prettify()

            divs=soup.select("div.grid-layout--cell")                 
            for idx, div in enumerate(divs):  
                user_name=div.select("div.user-details a")[0].get_text()
                location=div.select("span.user-location")[0].get_text()
                div_score=div.select("span.reputation-score")
                reputation=div_score[0].get('title')
                rep_score=reputation.split(" ")[2]
                rep_score="".join(rep_score.split(","))
                rep_score=int(rep_score)
                
                if(user_name):
                    usernames.append(user_name)
                else:
                    usernames.append(np.nan)
                    
                if(location):
                    locations.append(location)
                else:
                    locations.append(np.nan)
                    
                if(rep_score):
                    reputation_score.append(rep_score)
                else:
                    reputation_score.append(np.nan)
               
            next_page=soup.select("div.pager a")

            for idx, item in enumerate(next_page): 
                if item.select("span.page-numbers")[0].get_text()==" next":
                    link=item.get('href')
                    break
                else:
                    continue
            if not link:       
                break         
            page_url = urljoin("https://stackoverflow.com",link)
    
    usernamesdf = pd.DataFrame(usernames)
    locationsdf = pd.DataFrame(locations)
    reputationdf = pd.DataFrame(reputation_score)
    dataframe = pd.concat([usernamesdf, locationsdf, reputationdf], axis=1, sort=False)
    
    dataframe.columns = ['User_Name', 'Location', 'Reputation_Score']
    dataframe.to_csv("users_dataset.csv", sep=',', index=False, encoding='utf-8') 
    print("User Profile data saved in users_dataset.csv in the current working directory\n")
    return dataframe

if __name__=="__main__":
    user_data=getUserData()
```

### 8. Question, Answer and Comment Scraping Script

* Scraping 1000 top voted question and answers for categories (*android, java, javascript, php, python*)
* Creating android_dataset.csv, java_dataset.csv, javascript_dataset.csv, php_dataset.csv, python_dataset.csv

```python
# Function to get the Full Data
def getFullData(tag):
    data=[]
    allquestions = []
    allvotes = []
    allviews = []
    alltags = []
    alltimestamps = []
    allans=[]
    allurls=[]
    
    for i in range(1,21):
        url ="https://stackoverflow.com/questions/tagged/"+tag+"?page="+str(i)+"&sort=votes&pagesize=50"
        #print(url)
        
        try:
            page = urlopen(url)
        except Exception as e:
            print(e)
                
        soup = BeautifulSoup(page, "lxml")
        #print(soup)
        
# Block to get questions data
        questions = soup.select('div.summary h3 a')
        #questions = soup.find_all('a',attrs={'class':'question-hyperlink'})
        for question in questions:
            #print(question.get_text())
            allquestions.append(question.get_text())
        #print("Number of questions: ",len(allquestions))
        

# Block to get Answers data

        for question in questions:
            ans=''
            que = question.get_text()
            #print('\n',que)
            link=question.get('href')
            ans_url="https://stackoverflow.com"+link
            #print(ans_url,'\n')
            
            #Sends send a get request to the web page
            ans_page=requests.get(ans_url)
            #Status_code 200 indicates success 
            if ans_page.status_code==200:
                ans_page.content                      
                
                #Initiates a beautifulsoup object using the html source and Python’s html.parser
                ans_soup=BeautifulSoup(ans_page.content, 'html.parser')
                
                divs=ans_soup.select('div#answers div.post-text')
                #print(divs)
                
                for idx,div in enumerate(divs):
                    if idx==0:
                        #print("question number :",question)
                        answer=div.find_all('p')
                        #print(answer)
                        for i,d in enumerate(answer):
                            dn=d.get_text()
                            ans=ans+dn
                            #ans.append(d.get_text())
                            #ans="".join(ans.split(","))
                        
                    else:
                        break
            #print(ans)
            allurls.append(ans_url)
            allans.append(ans)
            
            
        # Block to get Votes data
        #votes = soup.find_all('div.vote-count-post high-scored-post strong')
        votesparent = soup.find_all('div',attrs={'class':'statscontainer'})
        for each in votesparent:
            vote = each.find('span',attrs={'class':'vote-count-post high-scored-post'})
            vote1 = each.find('span',attrs={'class':'vote-count-post'})
            if (vote):
                allvotes.append(vote.get_text())
            elif(vote1):
                allvotes.append(vote1.get_text())
            else:
                allvotes.append('NA')
        #print("Number of votes: ", len(allvotes))
    
    
        # Block to get Views data
        viewsparent = soup.find_all('div',attrs={'class':'statscontainer'})
        for each in viewsparent:
            view = each.find('div',attrs={'class':'views supernova'})
            if (view):
                number = int(''.join(view['title'].split(' ')[0].split(',')))
                allviews.append(number)
            else:
                allviews.append("NA")
        #print("Number of views: ", len(allviews))
        
        # Block to get Tags data
        summary = soup.find_all('div',attrs={'class':'summary'})
        #print(summary)
        for each in summary:
            tags = each.find_all('a',attrs={'class':'post-tag'})
            #print(tags)
            tagunit =[]
            for tag1 in tags:
                #print(tag.get_text())
                tagunit.append(tag1.get_text())
            alltags.append(tuple(tagunit))
        #print("Number of tags: ", len(alltags))
        #print(alltags)
        
        # Block to get Timestamp data
        timeparent = soup.find_all('div',attrs={'class':'started fr'})
        for each in timeparent:
            timestamp = each.find('span',attrs={'class':'relativetime'})
            if (timestamp):
                stamp = timestamp['title'].split(' ')[0]
                #print(stamp)
                alltimestamps.append(stamp)
            else:
                alltimestamps.append("NA")
        #print("Number of timestamps: ", len(alltimestamps))
        #print(alltimestamps)
        
    questionsdf = pd.DataFrame(allquestions)
    votesdf = pd.DataFrame(allvotes)
    viewsdf = pd.DataFrame(allviews)
    datedf = pd.DataFrame(alltimestamps)
    urldf = pd.DataFrame(allurls)
    answersdf = pd.DataFrame(allans)
    
    tagdf = None
    for tag2 in alltags:
        tagstring = [','.join(tag2)]
        if (tagdf) is None:
            tagdf = pd.DataFrame(tagstring)
        else:
            tgs = pd.DataFrame(tagstring)
            frames = [tagdf,tgs]
            tagdf = pd.concat(frames,ignore_index=True)
    #print(tagdf)
    length = len(allquestions) +1
    qnumberdf = pd.DataFrame({'Question_Number': range(1, int(length))})
    
    
    dataframed = pd.concat([qnumberdf,questionsdf,votesdf,viewsdf,datedf,tagdf,urldf,answersdf], axis=1, sort=False)
    
    dataframed.columns = ['Question_Number','Question', 'Votes','Views','Date','Tags','URL','Best_voted_answer']
    dataframed.to_csv(tag+"_dataset.csv", sep=',', index=False, encoding='utf-8')
    
    return dataframed

if __name__=="__main__":
    pythonData = getFullData('python')
    javaData = getFullData('java')
    javascriptData = getFullData('javascript')
    csharpData = getFullData('android')
    php = getFullData('php')
    print("Questions and answers data saved in respective csv in the current working directory")
```

### 13. Text Analysis on Questions
Populating top three similar questions for top five most viewed questions. We have used TF_IDF matrix and similarity score to get the top three similar questions based on the question being analyzed.

```python
wordnet_lemmatizer = WordNetLemmatizer()

# Function to get wordnet tag
def get_wordnet_pos(pos_tag):
    
    # if pos tag starts with 'J'
    if pos_tag.startswith('J'):
        # return wordnet tag "ADJ"
        return wordnet.ADJ
    
    # if pos tag starts with 'V'
    elif pos_tag.startswith('V'):
        # return wordnet tag "VERB"
        return wordnet.VERB
    
    # if pos tag starts with 'N'
    elif pos_tag.startswith('N'):
        # return wordnet tag "NOUN"
        return wordnet.NOUN
    
    elif pos_tag.startswith('R'):
        return wordnet.ADV
    else:
        # be default, return wordnet tag "NOUN"
        return wordnet.NOUN


# Function to tokenize the text
def tokenize(text):
    text = text.lower()
    pattern=r'\w[\w\'_.@-]*\w'
    tokens = nltk.regexp_tokenize(text, pattern)
    tokens=[token.strip(string.punctuation) for token in tokens]
    tokens=[token.strip() for token in tokens if token.strip()!='']
    #print(len(tokens))                   
    #print(tokens)
    tagged_tokens= nltk.pos_tag(tokens)
    stop_words = stopwords.words('english')
    stop_words+=["they'll", "can't"]    
    lemmatized_words=[wordnet_lemmatizer.lemmatize(word, get_wordnet_pos(tag)) for (word, tag) in tagged_tokens if word not in stop_words and word not in string.punctuation]
    token_count=nltk.FreqDist(lemmatized_words)
    return token_count


# Function to find similar doc
def find_similar_doc(doc_id, docs):
    docs_tokens={idx:tokenize(doc) for idx,doc in enumerate(docs)}
    #print(docs_tokens)
    
    dtm=pd.DataFrame.from_dict(docs_tokens, orient="index" )
    dtm=dtm.fillna(0)
    
    tf=dtm.values
    doc_len=tf.sum(axis=1)
    tf=np.divide(tf.T, doc_len).T
    
    df=np.where(tf>0,1,0)
    
    idf=np.log(np.divide(len(docs),np.sum(df, axis=0)))+1
    smoothed_idf=np.log(np.divide(len(docs)+1, np.sum(df, axis=0)+1))+1    
    smoothed_tf_idf=normalize(tf*smoothed_idf)
    
    tf_idf=normalize(tf*idf)
    #print(tf_idf)
    
    similarity = 1-distance.squareform(distance.pdist(tf_idf, 'cosine'))
    #print(similarity)
   
    best_matching_doc_id = np.argsort(similarity)[:,::-1][int(doc_id),1:4]
    #print(best_matching_doc_id)
    #print(type(best_matching_doc_id))
    
    similarity_score = similarity[doc_id][best_matching_doc_id]
    
    return best_matching_doc_id, similarity_score
    
    
# Function to find top three similar question
def topThreeSimilarQuestions(filename):
    print('\nFile under analysis: ',filename)
    similardata=pd.read_csv(filename, header=0)
    questionslist = similardata["Question"].values.tolist()
    #print(questionslist)
    sorteddata = similardata.sort_values('Views', ascending=False)
    fivemostviewedquestions= sorteddata.head(5)
    docidlist = fivemostviewedquestions['Question_Number'].values.tolist()
    
    #print('five most viewed questions /n')
    
    #print(fivemostviewedquestions['Question_Number'])
    
    #for index, row in fivemostviewedquestions.iterrows():
    num1=1
    for docid in docidlist:
        
        print('\n',num1,'.','Question under analysis: ',questionslist[docid])
        
        x,y = find_similar_doc(docid, questionslist)
        print('Similar Question numbers and their Similarity Scores: ',x,y)
        print('Top 3 similar questions of question under analysis:')
        num2=1
        for each in x:
            print(num2,'.',questionslist[each])
            num2+=1
        num1+=1

    return None

if __name__=="__main__":
    print("Top 3 similar python question for top 5 most viewed questions")
    python = topThreeSimilarQuestions('python_dataset.csv')
    print("Top 3 similar php question for top 5 most viewed questions")
    php = topThreeSimilarQuestions('php_dataset.csv')
    print("Top 3 similar javascript question for top 5 most viewed questions")
    javascript = topThreeSimilarQuestions('javascript_dataset.csv')
    print("Top 3 similar java question for top 5 most viewed questions")
    java = topThreeSimilarQuestions('java_dataset.csv')
    print("Top 3 similar android question for top 5 most viewed questions")
    android = topThreeSimilarQuestions('android_dataset.csv')
```

### 15. Topic Modelling

```python
# Function to get Tfidf matrix
def tfidf(docs):
    # Getting list of token for the document
    docs_tokens={idx:tokenize(doc) \
             for idx,doc in enumerate(docs)}

    # Generating document-term matrix
    dtm=pd.DataFrame.from_dict(docs_tokens, orient="index" )
    dtm=dtm.fillna(0)
    
    # generating normalized term frequency (tf) matrix        
    tf=dtm.values
    doc_len=tf.sum(axis=1)
    tf=np.divide(tf.T, doc_len).T
    
    # generating idf
    df=np.where(tf>0,1,0)

    smoothed_idf=np.log(np.divide(len(docs)+1, np.sum(df, axis=0)+1))+1    
    smoothed_tf_idf=tf*smoothed_idf
    
    return smoothed_tf_idf


# Function to generate word cloud
def getWordCloud(tag):
    for i in tag:
    
        data=pd.read_csv("Clean_Data/"+i+"_clean.csv", header=0, encoding = "ISO-8859-1")


        q = list(data.Question)    
        
        #Analyzing data

        tf_vectorizer = CountVectorizer(max_df=0.90,min_df=5, stop_words='english')

        tf = tf_vectorizer.fit_transform(q)

        tf_feature_names = tf_vectorizer.get_feature_names()

        print(tf_feature_names[0:10])

        X_train, X_test = train_test_split(tf, test_size=0.1, random_state=0)

        num_topics = 4

        lda = LatentDirichletAllocation(n_components=num_topics, \
                                    max_iter=10,verbose=1,
                                    evaluate_every=1, n_jobs=1,
                                    random_state=0).fit(X_train)

        num_top_words=20

        for topic_idx, topic in enumerate(lda.components_):  
            print ("Topic %d:" % (topic_idx))
            words=[(tf_feature_names[i],topic[i]) for i in topic.argsort()[::-1][0:num_top_words]]
            print(words)
            print("\n")

        num_top_words=50
        f, axarr = plt.subplots(2, 2, figsize=(8, 8))

        for topic_idx, topic in enumerate(lda.components_):

            # create a word:count dictionary
            f={tf_feature_names[i]:topic[i] for i in topic.argsort()[::-1][0:num_top_words]}

            # generate wordcloud in subplots
            wordcloud = WordCloud(width=480, height=450, margin=0, background_color="black");
            _ = wordcloud.generate_from_frequencies(frequencies=f);

            _ = axarr[math.floor(topic_idx/2), topic_idx%2].imshow(wordcloud, interpolation="bilinear");
            _ = axarr[math.floor(topic_idx/2), topic_idx%2].set_title("Topic: "+str(topic_idx));
            _ = axarr[math.floor(topic_idx/2), topic_idx%2].axis('off')

        plt.tight_layout()
        plt.show()
        
if __name__ == "__main__":
    tag=['classification_dataset','javascript_dataset', 'java_dataset', 'python_dataset', 'php_dataset', 'android_dataset',\
         'javascript_unanswered', 'java_unanswered', 'python_unanswered', 'php_unanswered', 'android_unanswered']
    getWordCloud(tag)
```

### 18. Matching questions and answers and evaluating the model (precision is low)

```python
# Function to match questions and answers
def match_question_answer(tag, filename):
    print('\nReference filename: ',filename)
    reference=pd.read_csv(filename, header=0)
    q= list(reference.Question)
    a= list(reference.Best_voted_answer)
    tf_idf=tfidf(q + a)
    #print(tf_idf.shape)
    similarity=1-distance.cdist(tf_idf[0:len(q)], tf_idf[len(q):], 'cosine')
    matched_q_id=np.argmax(similarity, axis=1)
    best_sim=np.max(similarity, axis=1)

    result=list(zip(list(range(len(q))),matched_q_id, best_sim))
    #print(matched_q_id,best_sim)
    return result


# Function to evaluate the questions and answers pair
def evaluate(result):
    
    performance=[]
    for min_sim in np.arange(0.05, 0.6, 0.05):
        
        filtered_result = [item for item in result if item[2]>=min_sim]
    
        p=len([item for item in filtered_result if item[0]==item[1]])/len(filtered_result)
        r=len(filtered_result)/len(result)
        
        performance.append((min_sim, p, r))
        
    df=pd.DataFrame(performance, columns=["min_sim", 'precision','recall'])
    df=df.set_index("min_sim")
    df.plot(kind="line")
    plt.show()
    

if __name__ == "__main__":
    evaluate(match_question_answer('python', 'Clean_Data/python_dataset_clean.csv'))
    evaluate(match_question_answer('javascript', 'Clean_Data/javascript_dataset_clean.csv'))
    evaluate(match_question_answer('java', 'Clean_Data/java_dataset_clean.csv'))
    evaluate(match_question_answer('php', 'Clean_Data/php_dataset_clean.csv'))
    evaluate(match_question_answer('android', 'Clean_Data/android_dataset_clean.csv'))
```

```python
# Function to get best classification parameters
def classify(data_file):
             
    df=pd.read_csv(data_file, header=0)
    #df_test=pd.read_csv(test_file, header=0)



    train, test = train_test_split(df, test_size=0.25)
    
    # Getting best parameters
    text_clf = Pipeline([('tfidf', TfidfVectorizer()),
                    # ('clf', MultinomialNB())
                     ('clf', svm.LinearSVC())
                   ])

    parameters = {'tfidf__min_df':[1,2,3,5,10],
              'tfidf__stop_words':[None,"english"],
              'clf__C': [0.5,1.0,5.0]  }

    metric =  "f1_macro"

    gs_clf = GridSearchCV(text_clf, param_grid=parameters, scoring=metric, cv=6)

    gs_clf = gs_clf.fit(train["Question"], train["Label"])

    for param_name in sorted(parameters.keys()):
        print("%s: %r" % (param_name, gs_clf.best_params_[param_name]))

    print("best f1_macro:", gs_clf.best_score_)
    
    # using best parameters to train models
    
    tfidf_vect = TfidfVectorizer(stop_words=gs_clf.best_params_['tfidf__stop_words'],\
                                min_df=gs_clf.best_params_['tfidf__min_df']) 
    
    dtm= tfidf_vect.fit_transform(train["Question"])

    clf = svm.LinearSVC(C=gs_clf.best_params_['clf__C']).fit(dtm, train["Label"])
    
    # test model
    dtm_test=tfidf_vect.transform(test["Question"])
    predicted=clf.predict(dtm_test)

    labels=sorted(train["Label"].unique())

    print(classification_report(test["Label"], predicted, labels=labels))

if __name__=="__main__":
    classify("classification_dataset.csv")
```

```python
# Function to get classification using NB and SVM for varying CV folds
def getClassCVFold():
    python = pd.read_csv("python_dataset.csv",header=0)
    java = pd.read_csv("java_dataset.csv",header=0)
    php = pd.read_csv("php_dataset.csv",header=0)
    android = pd.read_csv("android_dataset.csv",header=0)
    javascript = pd.read_csv("javascript_dataset.csv",header=0)


    frames = [python, java, php,android,javascript]
    qa_dataset = pd.concat(frames)


    questions = qa_dataset["Question"].tolist()
    #print(questions)
    
    tags = qa_dataset["Tags"]
    #tags.head()
    
    labels =[]
    for tag in tags:
        labels.append(tag.split(",")[0])
    #print(labels)
    
    questionsdf = pd.DataFrame(questions)
    
    labeldf = pd.DataFrame(labels)
    #labeldf.head()

    classification_df = pd.concat([questionsdf,labeldf], axis=1, sort=False)
    classification_df.columns = ['Question', 'Label']
    #classification_df.head()

    #classification_df.to_csv("classification_dataset.csv", sep=',', index=False, encoding='utf-8')


    # Creating TF-IDF Matrix with stop words removed
    tfidf_vect = TfidfVectorizer(stop_words="english") 

    # Generating tfidf matrix
    dtm= tfidf_vect.fit_transform(classification_df["Question"])

    #print("type of dtm:", type(dtm))
    #print("size of tfidf matrix:", dtm.shape)


    MultinomialDB_Precision = []
    MultinomialDB_Recall = []
    MultinomialDB_F1score = []

    linearSVM_Precision = []
    linearSVM_Recall = []
    linearSVM_F1score = []
    folds = []


    metrics = ['precision_macro', 'recall_macro', \
               "f1_macro"]


    for i in range(3,51):
        # initiating an i fold cross validation MultinomialNB model
        clf = MultinomialNB()

        cv = cross_validate(clf, dtm, classification_df["Label"], \
                            scoring=metrics, cv=i)

        #print("Test data set average precision:")
        #print(np.mean(cv['test_precision_macro']))
        MultinomialDB_Precision.append(np.mean(cv['test_precision_macro']))
        #print("\nTest data set average recall:")
        #print(np.mean(cv['test_recall_macro']))
        MultinomialDB_Recall.append(np.mean(cv['test_recall_macro']))
        #print("\nTest data set average fscore:")
        #print(np.mean(cv['test_f1_macro']))
        MultinomialDB_F1score.append(np.mean(cv['test_f1_macro']))


        #initiate an i fold cross validation linear SVM model
        clf = svm.LinearSVC()

        cv = cross_validate(clf, dtm, classification_df["Label"], \
                            scoring=metrics, cv=i)

        #print(cv)

        #print("Test data set average precision:")
        #print(np.mean(cv['test_precision_macro']))
        linearSVM_Precision.append(np.mean(cv['test_precision_macro']))
        #print("\nTest data set average recall:")
        #print(np.mean(cv['test_recall_macro']))
        linearSVM_Recall.append(np.mean(cv['test_recall_macro']))
        #print("\nTest data set average fscore:")
        #print(np.mean(cv['test_f1_macro']))
        linearSVM_F1score.append(np.mean(cv['test_f1_macro']))

        folds.append(i)


    NBprecision = pd.DataFrame(MultinomialDB_Precision)
    NBrecall = pd.DataFrame(MultinomialDB_Recall)
    NBf1score = pd.DataFrame(MultinomialDB_F1score)

    SVMprecision = pd.DataFrame(linearSVM_Precision)
    SVMrecall = pd.DataFrame(linearSVM_Recall)
    SVMf1score = pd.DataFrame(linearSVM_F1score)
    folds = pd.DataFrame(folds)
    # SVMprecision.head()

    precision_df = pd.concat([NBprecision,SVMprecision,folds], axis=1, sort=False)
    precision_df.columns = ['MultinomilaNB', 'Linear SVM','Number of CV folds']
    # precision_df.head()

    recall_df = pd.concat([NBrecall,SVMrecall,folds], axis=1, sort=False)
    recall_df.columns = ['MultinomilaNB', 'Linear SVM','Number of CV folds']
    # recall_df.head()

    f1score_df = pd.concat([NBf1score,SVMf1score,folds], axis=1, sort=False)
    f1score_df.columns = ['MultinomilaNB', 'Linear SVM','Number of CV folds']
    # f1score_df.head()

    precision = precision_df.set_index('Number of CV folds').plot(figsize=(10,5),grid = True)
    precision.set_ylabel("Average Precision")
    precision.set_title("Precision - NB vs SVM")

    recall = recall_df.set_index('Number of CV folds').plot(figsize=(10,5),grid = True)
    recall.set_ylabel("Average Recall")
    recall.set_title("Recall - NB vs SVM")

    f1score = f1score_df.set_index('Number of CV folds').plot(figsize=(10,5),grid = True)
    f1score.set_ylabel("Average F1 Score")
    f1score.set_title("F1 Score - NB vs SVM")

    pylab.show()
    
    
if __name__=="__main__":
    getClassCVFold()
```

```python
# Function for k-means clustering
def kMeansCluster():
    df=pd.read_csv("classification_dataset.csv", header=0, encoding = "ISO-8859-1")

    # Getting unique labels
    print(df["Label"].unique())

    # Generating Tfidf matrix with min document frequency as 5
    tfidf_vect = TfidfVectorizer(stop_words="english",min_df=5) 

    dtm= tfidf_vect.fit_transform(df["Question"])
    #print (dtm.shape)

    # Setting number of clusters
    num_clusters=3

    # Initializing clustering model using cosine distance for 20 repetition with different initial centroids
    clusterer = KMeansClusterer(num_clusters, cosine_distance, repeats=20)

    # Assiging samples to cluster labels starting from 0
    clusters = clusterer.cluster(dtm.toarray(), assign_clusters=True)

    # Printing the cluster labels of the first 5 samples
    print(clusters[0:5])

    
    # Using clusterer.means() which contains the centroids,
    # where each row is a cluster and column is a feature (word)
    centroids=np.array(clusterer.means())

    # Using Argsort to sort the matrix in ascending order 
    # and returning locations of features before sorting ([:,::-1] for reversing the order)
    sorted_centroids = centroids.argsort()[:, ::-1] 

    # Using get_feature_names() to get mapping between feature(word) index and feature (word)
    voc_lookup= tfidf_vect.get_feature_names()

    for i in range(num_clusters):

        # Getting the words with top 20 tf-idf weight in the centroid
        top_words=[voc_lookup[word_index] for word_index in sorted_centroids[i, :20]]
        print("Cluster %d:\n %s " % (i, "; ".join(top_words)))

    # Creating "cluster" column in the dataframe
    df["cluster"]=clusters
    df.head()

    # Generating crosstab between clusters and true labels
    confusion=pd.crosstab(index=df.cluster, columns=df.Label)
    print(confusion)

    mapping = confusion.idxmax(axis=1)
    for idx, t in enumerate(mapping):
        print("Cluster {}: Topic {}".format(idx, t))

    predicted_target=[mapping[i] for i in clusters]

    print(metrics.classification_report(df["Label"], predicted_target))

if __name__=="__main__":
    kMeansCluster()
```

### 23. LDA clustering

```python
# Function for LDA clustering
def ldaClustering():
    df=pd.read_csv("classification_dataset.csv", header=0, encoding = "ISO-8859-1")

    tf_vectorizer = CountVectorizer(max_df=0.90, \
                    min_df=50, stop_words='english')
    tf = tf_vectorizer.fit_transform(df["Question"])

    # Using get_feature_names() to get all the words
    tf_feature_names = tf_vectorizer.get_feature_names()

    print(tf_feature_names[0:10])
    print(tf.shape)

    # Splitting dataset into train (90%) and test datasets (10%) where
    # the test sets will be used to evaluate proplexity of topic modeling
    X_train, X_test = train_test_split(\
                    tf, test_size=0.1, random_state=0)


    # Converting the gensim corpus from the sparse tf matrix
    corpus = gensim.matutils.Sparse2Corpus(X_train, documents_columns=False)

    # Creating the mapping between id and words
    id2word={idx:w for idx, w in enumerate(tf_vectorizer.get_feature_names())}

    # Create a gensim dictionary from the corpus of
    # a dictionary that contains the frequency of each words and
    # the mapping between ids and words
    dictionary = corpora.Dictionary.from_corpus(corpus, id2word=id2word)

    # Setting number of topics
    NUM_TOPICS = 4

    ldamodel = gensim.models.ldamodel.LdaModel(corpus, alpha='auto', num_topics = NUM_TOPICS, \
                                               id2word=id2word, iterations=15)

    topics = ldamodel.print_topics(num_words=20)
    for topic in topics:
        print(topic)

    lda_display = pyLDAvis.gensim.prepare(ldamodel, corpus, dictionary, sort_topics=False)
    #pyLDAvis.display(lda_display)

    test_corpus = gensim.matutils.Sparse2Corpus(X_test, documents_columns=False)
    predict = ldamodel.get_document_topics(test_corpus)
    print(list(predict)[0:5])
    
    return lda_display

    
if __name__=="__main__":
    lda_display = ldaClustering()
    pyLDAvis.display(lda_display)
```
